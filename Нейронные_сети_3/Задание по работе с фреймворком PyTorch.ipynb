
Задача 4. 

Напишите функцию forward_pass(X, w) (w0 входит в w) для одного нейрона (с сигмоидой) с помощью PyTorch


```python
def forward_pass(X, w):
    return torch.sigmoid(X @ w)
```

```python
X = torch.FloatTensor([[-5, 5], [2, 3], [1, -1]])
w = torch.FloatTensor([[-0.5], [2.5]])
result = forward_pass(X, w)
print('result: {}'.format(result))
```

4) Затем измените форму тензора b, чтобы она совпадала с формой тензора a, получите тензор c<br>
5) Переместите c на CPU, переместите a на CPU<br>
6) Оберните их в Variable()<br>
7) Объявите тензор L = torch.mean((c - a)**2) и посчитайте градиент L по c ( то есть ∂L∂c)<br>
8) Выведите градиент L по c<br>

```python
import torch
```

```python
a = torch.FloatTensor(2, 3, 4).uniform_().cuda()
b = torch.FloatTensor(1, 8, 3).uniform_().cuda()
a.is_cuda
```

```python
c = b.view(2, 3, 4)
```

```python
a = a.cpu()
c = c.cpu()
a.is_cuda
```

```python
from torch.autograd import Variable
```

```python
a = Variable(a, requires_grad=False)
c = Variable(c, requires_grad=True)
```

```python
L = torch.mean((c - a)**2)
# Когда мы вызываем L.backward(), он вычисляет градиент потерь w.r.t всех параметров в L, 
# которые имеют requires_grad = True, 
# и сохраняет их в атрибуте parameter.grad для каждого параметра.
L.backward()
```

```python
print(c.grad)
```
